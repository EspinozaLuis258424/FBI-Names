```python
csv_file = "names.csv" 
namescsv = pd.read_csv(csv_file)
namescsv.head()
namescsv.info()
```

```python
class_df = namescsv[['first_name', 'last_name']].copy()
class_df.head()
```

```python
class_df = class_df[:30]
```

```python
class_df.head()
```

```python
class_list = class_df.first_name
class_list.head()
```

```python
wanted_df = pd.DataFrame({})
wanted_df['Test_Name'] = class_list
wanted_df['Name'] = ''
wanted_df['Total Records'] = ''
wanted_df['Reward'] = ''
wanted_df['Description'] = ''
wanted_df['Url'] = ''

wanted_df.head()
```

```python
print('Scraping FBI Data')
row_count = 0

for index, row in wanted_df.iterrows():
    
    name = row['Test_Name']
    response = json.loads((requests.get('https://api.fbi.gov/@wanted?title=' + name)).content)
    row_count = row_count + 1
    
    if response["total"] == 0:
        
        print(f' Record {row_count} NOT FOUND for {name}')
        wanted_df.loc[index, 'Total Records'] = response['total']
            
    else:
        print(f' Processing Record {row_count} for {name}')
        wanted_df.loc[index, 'Name'] = response['items'][0]['title']
        wanted_df.loc[index, 'Total Records'] = response['total']
        wanted_df.loc[index, 'Reward'] = response['items'][0]['reward_text']
        wanted_df.loc[index, 'Description'] = response['items'][0]['description']
        wanted_df.loc[index, 'URL'] = response['items'][0]['url']

```
