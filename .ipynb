## Scraping Data from FBI Most Wanted API

```python
#Make Connection With FBI API
response = requests.get('https://api.fbi.gov/@wanted?pageSize=1000&page=1')
data = json.loads(response.content)
```

```python
#explore Data
pprint(data['items'][0])
print(data['total'])
print(data['items'][0]['title'])
```

```python
#Make Dataframe
wanted_df = pd.DataFrame({})
wanted_df['ID'] = [i for i in range(800)]
wanted_df['First_Name'] = ''
wanted_df['Name'] = ''
wanted_df['Sex'] = ''
wanted_df['Race'] = ''
wanted_df['Birth_Place'] = ''
wanted_df['Eye_Color'] = ''
wanted_df['Weight'] = ''
wanted_df['Reward'] = ''
wanted_df['Description'] = ''
wanted_df['Modified'] = ''


wanted_df.head()
```

```python
#Run Loop to gather data

print('**********Scraping FBI Data********')
#Make Counters
row_count = 0
page_count = 1
i = 0
total = 0


for index, row, in wanted_df.iterrows():
    
   #Through testing I found that there were 16 pages worth of data, we need to loop through them to get the goods.

    response = json.loads((requests.get('https://api.fbi.gov/@wanted?pageSize=1000&page=' + str(page_count)).content))
    row_count = row_count + 1

    #Break loop after aquiring all records
    
    if total == 777:
        
        print('********Done*******')
        break
        
    #There are only 50 records per page
    else:
        if i == 50:
            
            i = 0
            page_count = page_count + 1
            
            print(f'-----------STARTING NEW PAGE------------')
            
        else:
            #populate DataFrame
            
            print(f'Currently on page {page_count}')

            #wanted_df.loc[index, 'Birthday'] = response['items'][i]['dates_of_birth_used']
            #print(response['items'][i]['dates_of_birth_used'])

            wanted_df.loc[index, 'Modified'] = response['items'][i]['modified']
            #print(response['items'][i]['modified'])

            wanted_df.loc[index, 'Race'] = response['items'][i]['race']
            #print(response['items'][i]['race'])
            
            wanted_df.loc[index, 'Eye_Color'] = response['items'][i]['eyes']

            wanted_df.loc[index, 'Birth_Place'] = response['items'][i]['place_of_birth']
            #print(response['items'][i]['place_of_birth'])

            wanted_df.loc[index, 'Sex'] = response['items'][i]['sex']
            #print(response['items'][i]['sex'])

            wanted_df.loc[index, 'Weight'] = response['items'][i]['weight']
            #print(response['items'][i]['weight'])

            wanted_df.loc[index, 'Name'] = response['items'][i]['title']
            #print(response['items'][i]['title'])

            wanted_df.loc[index, 'Reward'] = response['items'][i]['reward_text']
            #print(response['items'][i]['reward_text'])

            wanted_df.loc[index, 'Description'] = response['items'][i]['description']
            #print(response['items'][i]['description'])

            wanted_df.loc[index, 'URL'] = response['items'][i]['url']
            #print(response['items'][i]['url'])

            page_count = int(response['page'])
            
            print(f'Processing Record {i}')
            print('----------------------------------')
            
            i = i + 1
            total = total + 1
```
